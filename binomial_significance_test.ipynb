```python
import numpy as np
import matplotlib.pyplot as plt

# Number of samples in each trial
n = 7200

# Number of trials
num_trials = 1000

# Store the proportions for each trial
proportions = []

for _ in range(num_trials):
    # Randomly draw 7200 binary samples with equal probability
    samples = np.random.choice([0, 1], size=n, p=[0.5, 0.5])
    
    # Calculate the proportion of 1s
    proportion = np.sum(samples) / n
    
    # Store the result
    proportions.append(proportion)

# Plot the histogram of proportions
plt.figure(figsize=(10, 6))
plt.hist(proportions, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(0.5, color='red', linestyle='dashed', linewidth=2, label='Expected value (0.5)')
plt.title('Distribution of Proportions from 100,000 Trials of 7200 Binary Samples')
plt.xlabel('Proportion of 1s')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Calculate some statistics
mean_prop = np.mean(proportions)
std_prop = np.std(proportions)
print(f"Mean proportion: {mean_prop:.6f}")
print(f"Standard deviation: {std_prop:.6f}")
print(f"Theoretical standard deviation: {np.sqrt(0.5 * 0.5 / n):.6f}")
```
